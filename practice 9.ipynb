```python
a=data1.isnull().sum() 

b=data1.isnull().sum()/len(data1)*100

b

data1=data1.drop(['county_name', 'search_type'], axis=1)

#data1

```

```python

# Az sayda NaN  deyeri olan kateqorik sutunlari her hansi bir deyer ile evez edin.
#(Meselen gender sutununda male, female ve boshluqlar varsa, hemin boshluqlari Unknown ve ya mode()[0] ile evez edin)

# Numerical sutunlarda olan NaN  deyerleri ise statistik deyer ile evez edin.
#(Meselen average,min,max ve basqalari)

data1['driver_gender'].fillna(value=data1['driver_gender'].mode()[0], inplace=True)
data1['driver_age_raw'].fillna(value=data1['driver_age_raw'].mean(), inplace=True)
data1['driver_age'].fillna(value=data1['driver_age'].mean(), inplace=True)
data1['driver_race'].fillna(value=data1['driver_race'].mode()[0], inplace=True)
data1['violation_raw'].fillna(value=data1['violation_raw'].mode()[0], inplace=True)
data1['is_arrested'].fillna(value=data1['is_arrested'].mean(), inplace=True)
data1['stop_duration'].fillna(value=data1['stop_duration'].mode()[0], inplace=True)
data1['violation'].fillna(value=data1['violation'].mode()[0], inplace=True)

data1





```
